Ordenar y filtrar los datos 

```python
df_alumnos_cursando = df_estudiantes[df_estudiantes["termino"].isnull() & (df_estudiantes["causa_salida"] == 1)]
print(df_alumnos_cursando.shape)
df_alumnos_cursando
```

```python
df_alumnos_pasados = df_estudiantes[df_estudiantes["termino"].notnull() & (df_estudiantes["causa_salida"] != 1)]
print(df_alumnos_pasados.shape)
df_alumnos_pasados
```
